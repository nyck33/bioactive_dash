```python
"""
get all links then follow all links and get tables
"""

import pandas as pd
from bs4 import BeautifulSoup
import requests
import lxml.html as lh
#from langdetect import detect


url = "http://www.knapsackfamily.com/DietNavi/top.php"

soup = BeautifulSoup(requests.get(url).content, 'html.parser')


links = soup.find_all('a')
links_arr = []

for link in soup.find_all('a', href=True):
    if link['href'].startswith('result2'):
        links_arr.append(link['href'])

for i in range(3):
    print(links_arr[i])

```

```python

#prepend to links
def prepend(arr, str):
    str += '{0}'
    arr = [str.format(i) for i in arr]

    return arr


#call prepend()
url_prefix = "http://www.knapsackfamily.com/DietNavi/"
full_links_arr = prepend(links_arr, url_prefix )

for i in range(5):
    print(full_links_arr[i])
```

```python
#visit links and get the cells I want:
all_soups = []
count = 0
#print('here')
for link in full_links_arr:
    #print(f'count: {count}\nlink: {link}\n')
    response = requests.get(link)

    soup_next = BeautifulSoup(response.content, 'html.parser')

    all_soups.append(soup_next)
    #count+=1
    #if count > 3:
     #   break


```

```python
diseases=[]
for soup in all_soups:
    disease = soup.find("font", {"class":"blue"})
    #print(disease.contents)
    diseases.append(disease.contents[0])
    
for d in diseases:
    print(d)
```

# Translation library import here and activate function cell

```python
from googletrans import Translator
from time import sleep
def getTranslate(text,**kwargs):
    translator = Translator(service_urls=[
        'translate.google.jp', 'translate.google.co.kr', 'translate.google.ca', 'translate.google.com'])
    result = None
    while result == None:
        try:
            result = translator.translate(text,**kwargs)
        except Exception as e:
            #print(e)
            translator = Translator()
            sleep(0.5)
            pass
    return result     
```

```python
import copy
from googletrans import Translator
translated = []
count = 0
for d in diseases:
    translation = getTranslate(d, src='ja', dest='en')
    translated.append(translation)
    print(translation)
```

```python
# get text attribute from translations
diseases_en = []
for t in translated:
    diseases_en.append(t.text)
for d in diseases_en:
    print(d)
assert len(all_soups) == len(diseases_en)
```

```python
#have a dict {a name='top': [[row of table], [row of table],...]}
import copy
diseases_rows_dict = {}
curr_rows = []
for i in range (len(all_soups)):
    for row in all_soups[i].select('table.list1 tr'):
        tds = [td.get_text(strip=True) for td in row.select('td, th')]
        curr_rows.append(tds)
    disease = diseases_en[i]
    diseases_rows_dict[disease] = copy.deepcopy(curr_rows)
    curr_rows.clear()

for k,v in diseases_rows_dict.items():
    print(f'{k}: {v}\n')
```

```python
for d, rows in diseases_rows_dict.items():
    if len(rows)<1:
        print(d)
```

```python
#bring food to front if elment[0] of row[0] is 代謝物名
new_row = []
new_rows = []
reordered_dict = {}
skipped = []
for d, rows in diseases_rows_dict.items():
    disease = d
    if(len(rows)>0):
        headers = rows[0]
        if headers[0] == '代謝物名':
            for i in range(0, len(rows)):
                if i ==0:
                    new_rows.append(['food', 'en_name', 'healthEffect', 'metabollite', 'citation'])
                else:
                    curr_row = rows[i]
                    new_rows.append([curr_row[3], curr_row[1], curr_row[2], curr_row[0], curr_row[4]])
        else:
            for i in range(0, len(rows)):
                if i ==0:
                    new_rows.append(['food', 'en_name', 'healthEffect', 'metabollite', 'citation'])
                else:
                    new_rows.append(rows[i])
        reordered_dict[disease] = copy.deepcopy(new_rows)
        new_rows.clear()
        
    else:
        skipped.append(d)

for s in skipped:
    print(f'skip: {s}\n')
for k,v in reordered_dict.items():
    print(f'{k}: {v}\n')
```

```python
#key is disease, row[0] is col headers and all others are col values
#want list of dicts
#disease is now a column header
#['food', 'en_name', 'healthEffect', 'metabollite', 'citation']
disease_dicts_arr = []
disease_dict = {}
count = 0
for d, rows, in reordered_dict.items():        
    disease_dict['disease'] = str(d)
    for i in range(len(rows)):
        if i ==0:
            continue
        curr_row = rows[i]
        disease_dict['food']=str(curr_row[0])
        disease_dict['en_name']=str(curr_row[1])
        disease_dict['healthEffect']=str(curr_row[2])
        disease_dict['metabollite']=str(curr_row[3])
        disease_dict['citation']=str(curr_row[4])
        
    disease_dicts_arr.append(copy.deepcopy(disease_dict))
    disease_dict.clear()
    
for d in disease_dicts_arr:
    for k,v in d.items():
        print(f'{k}: {v}\n')
        
    
                            
```

```python
diseases_df = pd.DataFrame(disease_dicts_arr)
print(diseases_df.head())
print(diseases_df.columns)
```
