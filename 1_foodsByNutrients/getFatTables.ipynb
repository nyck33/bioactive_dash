```python
"""
get all links then follow all links and get tables
"""

import pandas as pd
from bs4 import BeautifulSoup
import requests
import lxml.html as lh
#from langdetect import detect


url = "https://www.webmd.com/diet/features/estimated-calorie-requirement"
url = "https://health.gov/our-work/food-nutrition/2015-2020-dietary-guidelines/guidelines/appendix-2"
soup = BeautifulSoup(requests.get(url).content, 'html.parser')

tables = soup.find_all('table')
tables_arr = []


#print(soup)
print(tables[0])
```

```python
rows1 = []
rows2 = []
i = 0
for table in tables:
    for row in table.select('table tr'):
        tds = [td.get_text(strip=True) for td 
               in row.select('td, th')]
        if i ==0:
            rows1.append(tds)
        else:
            rows2.append(tds)
    i+=1
    
print(rows1, "\n", rows2)
    
```

```python
# first elem is columns
cols = rows1[0]
cols = ['age', 'sedentary', 'moderately_active', 'active']
```

```python
males_dict = []
for i in range (1, len(rows1)):
    males_dict.append({cols[0]: rows1[i][0],
                            cols[1]: rows1[i][1],
                            cols[2]: rows1[i][2],
                            cols[3]: rows1[i][3]})

males_df = pd.DataFrame(males_dict)

males_df
   
```

```python
m_df = pd.DataFrame.from_dict(males_dict)
m_df
```
