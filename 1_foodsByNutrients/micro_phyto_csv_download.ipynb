```python
from bs4 import BeautifulSoup
import requests
from urllib.parse import urljoin
from os.path import basename


# base url for csv links
base="https://www.nal.usda.gov/sites/www.nal.usda.gov/files/"
url = requests.get("https://www.nal.usda.gov/fnic/nutrient-lists-standard-reference-legacy-2018").text
soup = BeautifulSoup(url)
# check the urls for csv links
links_arr = []
for a in soup.select("a[href]"):
    if 'csv' in a:
        print(a["href"])
        link = a["href"]
        
        links_arr.append(link)
        req = requests.get(link)
        url_content = req.text
        #decoded_content = url_content.decode('utf-8')
        
        csv_name = basename(link)
        file_path = "nutrient_name_csvs/" + csv_name
        with open(file_path, 'w') as f:
            f.writelines(url_content)
        
        f.close()
        #print(url_content)
        
#for link in (urljoin())
```

```python
import glob
file_names = glob.glob("nutrient_name_csvs/*.csv")
#print(file_names[:5])
nutrient_names_arr =  []
for name in file_names:
    csv_name = name.split("/")[1]
    nutrient = csv_name.replace(".csv", "")
    print(nutrient)
    nutrient_names_arr.append(nutrient)
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df_arr = []
for file in file_names:
    temp_df = pd.read_csv(file, skiprows=[0])
    df_arr.append(temp_df)
df_dict = dict(zip(nutrient_names_arr, df_arr))
```

```python
for k,v in df_dict.items():
    print(f'{k}\n: {v.head()}')
```

```python
#connect to mysql
from sqlalchemy import create_engine
engine = create_engine("mysql+pymysql://root:tennis33@localhost/dashcnf?charset=utf8mb4")
```

```python
test_df = df_dict['protein']
#print(test_df.columns.tolist())
#print(test_df.head())
test_df.to_sql('protein_foods', con=engine, if_exists='replace', chunksize=1000, index=False)
```

```python
for nutrient_name,df in df_dict.items():
    table_name = nutrient_name + "_foods"
    df.to_sql(table_name, con=engine, if_exists='replace', chunksize=1000, index=False)
```

```python
#connect to mysql
from sqlalchemy import create_engine
engine = create_engine("mysql+pymysql://root:tennis33@localhost/dashcnf?charset=utf8mb4")
```

```python
#upload all phytonutrient tables? or select few?
import glob
import pandas as pd
file_names = glob.glob("usdaDBFlavonoidContentSelFoods/*.csv")
df_arr = []
for name in file_names:
    df = pd.read_csv(name)
    #print(f'{df.columns}:\n{df.head()}')
    df_arr.append(df)
    
#get only the filename from path
csv_names_arr = []
for file in file_names:
    csv_name = file.split("/")[1]
    print(csv_name)
    csv_names_arr.append(csv_name)


df_dict = dict(zip(csv_names_arr, df_arr))


    
```

```python
print(len(df_dict))
for k,v in df_dict.items():
    print(f'{k}: {v.columns}')
```
